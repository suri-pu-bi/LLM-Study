### Bag of Words


1-1. scratch로 Bag of Words 구현




```python
documents = ['Hello, how are you',
             'Win money, win from home.',
             'Call me now',
             'Hello, Call me to say how you are?']

# 대문자를 소문자로 변환
lower_case_documents = []
for i in documents:
    lower_case_documents.append(i.lower())
print(lower_case_documents)
```

```python
# 구두점 삭제
sans_punctuation_documents = []
import string

for i in lower_case_documents:
    # 문자가 구두점에 속하지 않을 경우만 문자 유지, 구두점일 경우 제외됨
    sans_punctuation_documents.append(''.join(c for c in i if c not in string.punctuation))

print(sans_punctuation_documents)
```

```python
# 전처리, 단어별로 분리 / 분리가 되어야 몇번 등장했는지 카운팅 가능
preprocessed_documents = []
for i in sans_punctuation_documents:
    preprocessed_documents.append(i.split(' '))

print(preprocessed_documents)
```

```python
# Bag of Words
frequency_list = []
import pprint # pretty print
from collections import Counter

for i in preprocessed_documents:
    frequency_list.append(Counter(i))

pprint.pprint(frequency_list)
```

1-2. scikit-Learn으로 Bag of Words 구현

```python
import pandas as pd
documents = ['Hello, how are you',
             'Win money, win from home.',
             'Call me now',
             'Hello, Call me to say how you are?']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vector = CountVectorizer()
count_vector.fit(documents)
```

```python
count_vector.get_feature_names_out()
```

```python
# Bag of Words
doc_array = count_vector.transform(documents).toarray()
doc_array
```

```python
frequency_matrix = pd.DataFrame(doc_array, index=documents, columns=count_vector.get_feature_names_out())
frequency_matrix
```
